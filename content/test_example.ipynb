Before we begin, we'll state some notation:
* $P_0$ is a distribution on a space $\mathcal{Z}$. The authors call this the _data generating distribution_. 
* $Z$ is our training data, which is sampled from this distribution.
* $l$ is the loss function we aim to minimize. The loss function is parametrized by $\theta$ and written as $l(\theta; Z)$. 
* $\mathcal{P}$ is a family of distributions "close to" $P_0$. We will formalize our notion of "close to" shortly.

We can now state our objective below.
$$\min_{\theta \in \Theta} \sup_{P \in \mathcal{P}} \mathbb{E}_{P}\left[l(\theta; Z)\right]$$
Intuitively, we are minimizing the worst case loss over distributions close to the one our training data was sampled from. 

Our next step is to characterize $\mathcal{P}$. To do so, the authors use a $\rho$-Wasserstein ball centered at $P_0$. Specifically, we have:
$$\mathcal{P} = \left\{P:W(P, P_0) \le \rho\right\}$$
